## Initialize Libraries needed

```python
newsapi = NewsApiClient(api_key = my_api_key)
```

## Initialize Data
This section shows that the data will be coming from techcrunch.com, and only articles from March 11 to 12, 2021 will be extracted. The News API of python has a limit of 100 being the maximum number assigned for page_size, which means that only 100 articles will be extracted.

```python
data = newsapi.get_everything(domains="techcrunch.com", 
                              language='en', 
                              page_size=100, 
                              from_param="2021-03-11", 
                              to="2021-03-12")
```

Focusing on article datas only, and checking the date of each.

```python
for x, y in enumerate(articles):
    print(f'{x}  {y["publishedAt"]}')
```

## Export as json

```python
import json
news_json = []

for x, y in enumerate(articles):
    date=y['publishedAt']
    title=y['title']
    full_article=y['content']
    author=y['author']
    news_json.append({
        "date": date,
        "title": title,
        "full article": full_article,
        "author": author
    })

news_json
with open('news.json', 'w') as outfile:
    json.dump(news_json, outfile)
```
